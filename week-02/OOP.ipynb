```python
class Group:
  count = 0

  def __init__(self, name, debut, members): # (처음에는 무조건 self를 넣어야함)
    self.name = name
    self.debut = debut
    self.members = members
    Group.count += 1

  def printGroup(self):
    print(self.name, self.debut, self.members)

  @classmethod
  def printCount(cls):
    print(cls.count)

p1 = Group("아이브", 2021, ["유진", "원영", "이서"])
p1.printGroup()
p1.printCount()
p2 = Group("에스파", 2019, ["윈터","닝닝","카리나"])
p2.printGroup()
p2.printCount()
p3 = Group("블랙핑크", 2019, ["제니","지수","로제"])
p3.printGroup()
p3.printCount()
print(Group.count)
```

```python
# Group 클래스를 상속받는 GirlGroup 새로운 클래스 정의

class GirlGroup(Group): # (superclass 이름)
  def __init__(self, name, debut, members, leader):
    super().__init__(name, debut, members)
    self.leader = leader
  def printGroup(self):
    print(self.name, self.debut, self.members, self.leader)

p2 = GirlGroup("에스파", 2019, ["카리나","윈터","지젤", "닝닝"], "한승철")
p2.printGroup()
```

```python
#UML
# 부모 클래스 (Super Class)
class Animal:
  def __init__(self, name):
    self.name = name # 공통 속성

  def printName(self):
    print(self.name)

  def sing(self):
    return "하하핳"

# 자식 클래스 (Sub Class) - Animal을 상속받음
class Dog(Animal):
  def __init__(self, name):
    super().__init__(name)

  def sing(self): # 부모 메서드를 재정의
    print(self.name+"멍멍")

class Cat(Animal):
  def __init__(self, name):
    super().__init__(name)

  def sing(self):
    print(self.name+"야옹")

# 객체 생성
dog = Dog("멍멍이")
cat = Cat("냥냥이")

dog.printName()
dog.sing()

cat.printName()
cat.sing()
```

```python
class Mammal:
  def birth(self):
    print("새끼를 낳는다")

class Fish:
  def swim(self):
    print("헤엄친다.")

# 다중 상속 (Animal, Swimmer 둘 다 상속)
class Dolphin(Mammal, Fish):
  pass

d = Dolphin()
d.birth()
d.swim()
```

```python
# project 2

class Character:
    def __init__(self, userID):
        self.userID = userID
        self.weapon = "맨손"
        self.health = 0

    def getWeapon(self, weapon):
        self.weapon = weapon

    def plusHealth(self, health):
        self.health += health

    def minusHealth(self, health):
        self.health -= health

    def printCharacter(self):
        print(f"ID: {self.userID} | 무기: {self.weapon} | 체력: {self.health}")
```

```python
class Warrior(Character):
    def __init__(self, userID, name, strengthHit):
        super().__init__(userID)
        self.name = name
        self.strengthHit = strengthHit

    def hit(self):
        print(f"[{self.name}] 타격 스킬, 피해량: {self.strengthHit}")
```

```python
class Magician(Character):
    def __init__(self, userID, name, strengthHeal):
        super().__init__(userID)
        self.name = name
        self.strengthHeal = strengthHeal

    def heal(self):
        print(f"[{self.name}] 치유 스킬, 회복량: {self.strengthHeal}")
```

```python
class Archer(Character):
    def __init__(self, userID, name, strengthShot):
        super().__init__(userID)
        self.name = name
        self.strengthShot = strengthShot

    def shot(self):
        print(f"[{self.name}] 사격 스킬, 피해량: {self.strengthShot}")
```

```python
w1 = Warrior(userID="qkr_sj", name="박성준", strengthHit=400)

w1.plusHealth(1000)
w1.getWeapon("무한의 대검")

w1.printCharacter()
w1.hit()

print(" 몬스터에게 공격당함 ")
w1.minusHealth(200)
w1.printCharacter()
```

```python
# Project 3
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printPerson(self):
        print(f"이름: {self.name}, 나이: {self.age}")
```

```python
class Doctor(Person):
    def __init__(self, name, age, specialty, hourly_rate):
        super().__init__(name, age)
        self.specialty = specialty
        self.hourlyRate = hourly_rate

    def printDoctor(self):
        super().printPerson()
        print(f"진료 분야: {self.specialty}, 시간당 진료비: {self.hourlyRate}원")

    def diagnose(self, condition_name):
        print(f"[{self.name}] 의사가 '{condition_name}'에 대한 진단을 내림")
```

```python
class Patient(Person):
    def __init__(self, name, age, patient_id, consultation_time):
        super().__init__(name, age)
        self.patientID = patient_id
        self.consultationTime = consultation_time

    def printPatient(self):
        super().printPerson()
        print(f"환자 번호: {self.patientID}, 예상 진료 시간: {self.consultationTime}분")

    def describeSymptoms(self, symptoms_list):
        print(f"[{self.name}] 환자가 다음 증상을 설명함: {', '.join(symptoms_list)}")
```

```python
class Care:
    def __init__(self, care_id, doctor_object, patient_object):
        self.careID = care_id
        self.doctor = doctor_object
        self.patient = patient_object

    def getCareInfo(self):
        print(f"=== 진료 기록 (번호: {self.careID}) ===")
        print("[담당 의사 정보]")
        self.doctor.printDoctor()
        print("[내원 환자 정보]")
        self.patient.printPatient()
        print()
```

```python
doctor1 = Doctor(name="박성준", age=45, specialty="내과", hourly_rate=100000)
patient1 = Patient(name="공준성", age=32, patient_id="010101", consultation_time=30)

care1 = Care(care_id="A1", doctor_object=doctor1, patient_object=patient1)

care1.getCareInfo()

print("\n--- 추가 행동 ---")
doctor1.diagnose("감기")
patient1.describeSymptoms(["기침", "구토", "발열"])
```
