# Matplotlib
# 데이터를 그래프로 그려주는 라이브러리
# 학습 데이터 분석, 전처리 / 학습,모델 성능 검증

```python
import matplotlib.pyplot as plt
```

```python
# plot
# 데이터를 선으로 시각화
# 다음에서 data는 y값, x축은 자동 생성
import matplotlib.pyplot as plt
data = [1,2,3,4,5]
plt.plot(data)
plt.title('기본 선 그래프')
plt.show()
```

```python
categories = ['A','B','C','D']
values = [4,6,1,8]
plt.bar(categories,values)
plt.title('막대 그래프 예시')
plt.show()
```

```python
days = [1,2,3] # 일자별 x축
az = [2,4,8]  # 아스트라 접종 수
pfiza = [5,1,3] # 화이자 접종 수
moderna = [1,2,5] # 모더나 접종 수

plt.plot(days,az, label ='az', marker='v')  # day 별 az접종자 수 & 범례추가 & 마커 변경 / 색 변경은 color
plt.plot(days,moderna, label='moderna', marker='o',linestyle='--')
plt.plot(days,pfiza, label='pfiza', marker='s')
plt.legend()
# 머신러닝 학습이 제대로 되고있는지 확인하는 최고의 방법은 그래프를 그려보는 것이다.
```

```python
# Scatter
# 사용 데이터 설정
import numpy as np
x = 4 + np.random.normal(0,2,24)
y = 4 + np.random.normal(0,2,len(x))
# size and color:
sizes = np.random.uniform(15,80,len(x))
colors = np.random.uniform(15,80,len(x))

#방법1
plt.scatter(x,y, s=sizes, c=colors, vmin=0, vmax=100)
```

```python
# Bar
# 위에 두개에 비하면 많이 쓰이진않음 / 비교할 때 많이 사용함
x = 0.5 + np.arange(8)
y = np.random.uniform(2,7,len(x))

# 방법1
plt.bar(x,y, width=1, edgecolor = "white", linewidth=0.7)
```

```python
# 여러개의 plot을 그리는 방법
# subplot(행개수,열개수,plot 번호)
data1 = np.arange(1,50) # 1부터 49까지의 데이터
plt.subplot(2,1,1) # 그래프를 2행 1열로 나눈 뒤, 첫 번째 영역(위쪽)에 plot을 그리겠다는 의미
plt.plot(data1) # data1을 첫 번째 영역에 plot

data2 = np.arange(50,100)
plt.subplot(2,1,2) # 두 번째 영역(아래쪽)에 plot을 그리겠다는 의미
plt.plot(data2) # data2를 두번째 영역에 plot

plt.show()
```
