```python
# numpy
# np.array()로 생성(list, tuple 넣는다)
# ML 에러, 폭망 주요 원인  a,ndim : 차원 / a.shape : 모양 / a.dtype : 자료형

import numpy as np

data1 = [1,2,3,4,5,6]
data2 = [{10,20,30},{40,50,60}]

a = np.array(data1)
print(data1)
print(a)
print(a.shape)
# Python 문법상 요소가 1개인 튜플은 반드시 뒤에 ,를 포함해야 합니다
# 6개의 칸을 가진 1차원 공간의 형태
print(a.ndim)
print(a.dtype)
print(a*-1)
print(a+a)

b = np.array(data2)
print(b)
print(b.shape)
print(b.ndim)
print(b.dtype)
```

```python
np.zeros(10)    # 0으로 이루어진 array
np.ones((2,3))    # 1로 이루어진 array
np.arange(10)   # range 0~9
arr = np.array([1.2, 3.4, -0.9, -2.1, 100.])
arr_int = arr.astype(np.int32)
print(arr_int)
num_str = np.array(['1.234', '-2.29', '1000'], dtype = np.bytes_)
num = num_str.astype(float)
print(num)
print(num.dtype)
```

```python
data1 = np.array([[1,2,3],[4,5,6]])
data2 = np.array([[1,2],[3,4],[5,6]])

print(data1)
print(data2)
print(np.dot(data1,data2))
```

```python
data1 = np.array([[1,2,3],[4,5,6]])

print(data1)
print(np.sum(data1))
print(np.sum(data1, axis = 0)) # axis 0: 행(row) 방향으로 동작 (위에서 아래로).
print(np.sum(data1, axis = 1)) # axis 1: 열(column) 방향으로 동작 (왼쪽에서 오른쪽으로)

print(np.mean(data1))
print(np.mean(data1, axis = 0))
print(np.mean(data1, axis = 1))

print(np.min(data1))
print(np.min(data1, axis = 0))
print(np.min(data1, axis = 1))

# argmax 가장 많이 쓰임, 생성형 AI
print(np.argmax(data1))
print(np.argmax(data1, axis = 0))  # axis=0: 각 열에서 가장 큰 값의 행 번호를 찾음
print(np.argmax(data1, axis = 1))  # axis=1: 각 행에서 가장 큰 값의 열 번호를 찾음
```

```python
# flatten 매우 중요
# EX) 차량 번호판, 간단한 이미지 인식
data1 = [1,2,3,4,5,6]
a = np.array(data1)
print(a)
a = a.reshape(2,3)
print(a)
a = a.flatten()
print(a)
```

```python
# 합치기
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate((arr1,arr2))
print(arr)
```

```python
arr1 = np.array(([1,2],[3,4]))
arr2 = np.array(([5,6],[7,8]))
arr = np.concatenate((arr1,arr2), axis=1)
print(arr)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.stack((arr1,arr2), axis=1)
print(arr)
```
