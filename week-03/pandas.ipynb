```python
from pandas import DataFrame
import pandas as pd # Corrected import statement for pandas
FILE_PATH = '/content/' # Define FILE_PATH
filename = '0325.csv'
df = pd.read_csv(FILE_PATH+filename)

print(df.shape)
print(df.columns)
print(df.index)
```

```python
# column, index 이름 변경
df.columns = ['group', 'member', 'company', 'year']
# df.index = ['one', 'two', 'three', 'four', 'five']

df
```

```python
# Column data 추출
# 원하는 column들 뽑아서 새로운 data frame에 저장
df_g = df['group']
df_g

# 복수의 column
df_1 = df[['member', 'year']]
df_1
```

```python
# 조건으로 column 데이터 추출

print( df['company'] == 'sm')
print( df['year'] > 2020)
print( (df['member'] >= 5) & df['year'] == 2021)
```

```python
# row data 추출
print(df[1:3])
print(df[:3]) # ex) 1~3번째
print(df[3:])
```

```python
# 원하는 row들이 불연속인 경우
# loc, iloc 사용한다 loc는 location, iloc는 index location
df0 = df.loc[[0,2]]
df0
```

```python
# 원하는 column, row 데이터 추출
df.colums = [0,1,2,3]
print(df.loc[1:3. :2])
print(df.loc[:.1:3])
```

```python
# Groupby 그룹화 / mean 평균
df0 = df.groupby('member')
df0['year'].mean()
```

```python
# row 삭제
# drop() 사용
# 복수 삭제시엔 리스트로 한다

print(df)
df = df.drop(df.index[2])
print(df)
```

```python
print(df)
df = df.drop(df.index[[1,4]])
print(df)
```

```python
# column 삭제
# drop (,axis=1) 사용
print(df)
df = df.drop('member', axis=1)
print(df)
```
